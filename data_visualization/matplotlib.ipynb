# Matplotlib

```python
# Matplotlib is a plotting library for the Python programming language and its numerical mathematics extension NumPy. used for 2D graphics.
import matplotlib.pyplot as plt
x = [1,2,3,4,5]
y = [2,3,5,8,11]

# Create a simple line plot
plt.plot(x, y)
plt.xlabel('X-axis Label')
plt.ylabel('Y-axis Label') 
plt.title('Simple Line Plot')
plt.show()
```

```python
x = [1,2,3,4,5]
y = [2,3,5,8,11]
# Create the Customized line plot
plt.plot(x,y,color = 'Yellow',linestyle = "--",marker ='o',linewidth = 3,markersize = 12,markerfacecolor = 'Red',markeredgewidth = 3,markeredgecolor = 'Blue')
plt.grid(True)
```

##### Multiple Plots

```python
# Sample Data
x = [1, 2, 3, 4, 5]     
y1 = [2, 3, 5, 8, 11]
y2 = [1, 4, 6, 9, 12]

plt.figure(figsize=(6,4))
# this indicates size of the figure with width 6 and height 4

plt.subplot(2, 2, 1)
plt.plot(x, y1, color='Red')
plt.title('PLOT-1')

plt.subplot(2,2,2)
plt.plot(x, y2, color='Blue')
plt.title('PLOT-2')

plt.subplot(2,2,3)
plt.plot(x, y1, color='Green', linestyle='--')
plt.title('PLOT-3')

plt.subplot(2,2,4)
plt.plot(x, y2, color='Orange', marker='o')
plt.title('PLOT-4')

plt.show()  # Ensure the plots are displayed
#explanation: The code demonstrates how to create simple and customized line plots using Matplotlib in Python. It includes examples of setting colors, line styles, markers, and creating multiple subplots in a single figure.
# it contains rows and columns to arrange multiple plots in a grid layout.
# where (2,2,1) indicates 2 rows, 2 columns, and the first subplot position. And there will be total 4 subplots in the figure.
```

### BarPlot

```python
categories = ['A', 'B', 'C', 'D', 'E']
values = [5, 7, 3, 8, 4]

# Create a bar plot
plt.bar(categories,values,color = 'Blue')
plt.xlabel('Categories')
plt.ylabel('Values')
plt.title('Simple Bar Plot')
plt.show()
```

### Histograms

```python
#Histogram is used to reperesent the distribution of a numerical dataset by dividing the data into intervals (bins) and counting the number of data points that fall within each bin.
```

```python
#sample data
data = [12, 15, 14, 10, 18, 20, 22, 25, 30, 28, 35, 40, 38, 45, 50, 55, 60, 65, 70, 75]
# Create a histogram
plt.hist(data, bins=5, color='Black', edgecolor='Yellow')
```

### Scatter plot

```python
# Sample Data
x = [1, 2, 3, 4, 5]
y = [2, 3, 5, 8, 11]
plt.scatter(x,y,color='Purple',marker ='<')
```

### Pie Chart


```python
labels = ['A', 'B', 'C', 'D', 'E']
sizes = [15, 30, 45, 10, 20]
colors = ['Gold', 'Lightblue', 'Lightgreen', 'Pink', 'Orange']
explode = (0.1, 0, 0, 0, 0)  # move the 1st slice slightly outwards
# Create a pie chart
plt.pie(sizes, labels=labels, colors=colors, explode=explode, autopct='%1.1f%%', shadow=True, startangle=140)
# autopct is used to display percentage value on each slice of pie chart
# shadow adds a shadow effect to the pie chart
# startangle rotates the start of the pie chart by given angle
# explode is used to separate a slice from the pie chart for emphasis
```

### Sales Data Visualization

```python
import pandas as pd
sales_data_df = pd.read_csv('sales_csv.csv')
sales_data_df.head(5)
```

```python
# plot total sales by Product
total_sales_by_product = sales_data_df.groupby('Product')['Price'].sum()
total_sales_by_product.plot(kind='bar', color='Cyan', title='Total Sales by Product')
```

```python
#plot sales trend over time
sales_trend = sales_data_df.groupby('Date')['Total'].sum().reset_index()
plt.plot(sales_trend['Date'], sales_trend['Total'], marker='o', color='Magenta')
```
